```python
import pandas as pd

df_models = pd.read_csv('django_models.csv')
df_views = pd.read_csv('django_making_queries.csv')

df_models.columns = ['Topic', 'Content', 'Code']
df_views.columns = ['Topic', 'Content', 'Code']

final_df = pd.concat([df_models, df_views], ignore_index=True)
final_df.to_csv('django_data.csv', index=False, encoding='utf-8')

print(len(final_df))

print(final_df['Topic'].value_counts().head())
```
